```python
import numpy as np
from sys import exit
import matplotlib.pyplot as plt

archi="DatosEx.dat"

#leer el archivo DatosEx.dat 
try:
    archivo=open(archi,"r") #r-read w-write w+ read and write
except:
    exit("No se pudo leer el archivo {0} \n".format(archi))

#hacer arreglos vacios para x; y
x=[]
y=[]

#forma de leer los datos
for linea in archivo:
    linea = linea.strip()#separador espacio
    datos = linea.split()#separador espacio
    #datos en cada arreglo
    x += [float(datos[0])]
    y += [float(datos[1])]
    
archivo.close()

print "DATOS DEL ARCHIVO"
print x,y #imprime los datos de DatosEx.dat en x; y
```

```python

print "POSIBILIDAD DE RAICES"
raices = []
for i in range(len(y)-1):
    z=y[i]*y[i+1]
    if z<0:
        raiz = x[i]-y[i]*((x[i+1]-x[i])/(y[i+1]-y[i])) 
        raices += [raiz]
        print raiz #hay 10 posibles raices
    w=[y[i],y[i+1]]
    #print w
```

```python
#"GRÁFICA DE LOS DATOS DEL ARCHIVO CON INTERPOLACIÓN"
def lag_inter(x,x_val,y_val):
    k=len(x_val)
    suma = 0.0
    
    for j in range(k):
        p= 1.0
        for m in range(k):
            if m != j:
                p *= float((x-x_val[m])/(x_val[j]-x_val[m]))
        suma += p*y_val[j]
    return suma
    
n=len(x)
plt.plot(x,y,"yo")
for i in range(len(raices)):
    plt.plot(raices[i],0.0,"ro")

for i in range (n-1):
    x_nuevo = np.linspace(x[i],x[i+1],10) #da pasos de tamaño 10
    y_nuevo = np.empty(len(x_nuevo)) #hace un arreglo vacio
    for j in range(len(x_nuevo)): #llena el arreglo con la funcion anterior
        y_nuevo[j] = lag_inter(x_nuevo[j],x[i:i+3],y[i:i+3]) #cambiar i+3 para cambiar el grado del polinomio
    plt.plot(x_nuevo,y_nuevo,"b-")
    
plt.show()
```
